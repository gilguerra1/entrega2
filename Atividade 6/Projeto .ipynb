```python
from pymongo import MongoClient
import requests
```

```python
#Acessando a api do banco central para a estrair os dados
url = "https://olinda.bcb.gov.br/olinda/servico/Pix_DadosAbertos/versao/v1/odata/ChavesPix(Data=@Data)?@Data='2022-05-19'&$top=100&$format=json"

response = requests.get(url)
print(response)
json_selic = response.json()
```

```python
print(json_selic)
```

```python
#Trnaformando o arquivo json em uma lista para adicionar oa mongo
if not isinstance(json_selic, list):
    json_selic = [json_selic]
```

```python
#Estabelecendo conexão com o mongo e acessando a collection 
client = MongoClient("mongodb://localhost:27017")
db = client.json
collection = db.banco_central
```

```python
#Inserindo os dados no banco de dados
dados = collection.insert_many(json_selic)
print(dados)
```

```python
#Lendo os dados do banco de dados
cursor = collection.find()

for document in cursor:
    print(document)
```
